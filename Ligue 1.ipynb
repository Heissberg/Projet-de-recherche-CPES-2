```python
#encodage du resultat à la mi temps
data.loc[:,'HTR'] = data.loc[:,'HTR'].map({'H':1,'D':0,'A':-1})
```

```python
#Encodage du resultat de fin de match
data.loc[:,'FTR'] = data.loc[:,'FTR'].map({'H':1,'D':0,'A':-1})
```

```python
#délimitation de la moitié de la saison
cutoff = datetime.strptime('01/01/2019','%d/%m/%Y')
```

```python
#création d'une indicatrice si match appartient à premiere moitié saison
idx = [0 if datetime.strptime(data.iloc[k,0],'%d/%m/%Y') < cutoff else 1 for k in range(data.shape[0])]
```

```python
data.insert(1,'Test group',idx)
```

```python
data
```

```python
equipes = data.HomeTeam.unique()
```

```python
equipes
```

```python
n_matchs_MS = {}
for i in equipes:
    n_matchs_MS[i] = len(data[(data['HomeTeam']==i)&(data['Test group'] == 0)] + data[(data['AwayTeam']==i)&(data['Test group'] == 0)])
```

```python
n_matchs_MS
```

```python
#Création dictionnaire référençant ensemble des matchs du groupe train n'ayant pas été joué en premiere moitié de saison
matchs_manquants = {}
for i in equipes:
    if n_matchs_MS[i]<19:
        l = list(equipes)
        l.remove(i)
        for j in data[(data['HomeTeam']==i)&(data['Test group']==0)]['AwayTeam']:
            l.remove(j)
        for j in data[(data['AwayTeam']==i)&(data['Test group']==0)]['HomeTeam']:
            l.remove(j)
        matchs_manquants[i] = l
```

```python
matchs_manquants
```

```python
#Assignation des matchs du groupe train au groupe train
for i in matchs_manquants.keys():
    for j in matchs_manquants[i]:
        ref = data[(data['HomeTeam']==i) & (data['AwayTeam']==j)].index[0]
        data.iloc[ref,1] = 0
        matchs_manquants[i].remove(j)
        matchs_manquants[j].remove(i)
```

```python
# S = shots, ST = shots on target, C = corners
additional_stat_names = ['S', 'ST', 'C']

# décline en Home et Away
# t = taken
off_home_stat_names = ['t_FTHG', 't_HTHG'] + ['t_H' + stat_name for stat_name in additional_stat_names]
off_away_stat_names = ['t_FTAG', 't_HTAG'] + ['t_A' + stat_name for stat_name in additional_stat_names]
off_stat_names = off_home_stat_names +  off_away_stat_names
off_stat_names
```

```python
stats_automne = pd.DataFrame(index = equipes, columns = off_stat_names)

# pour chaque équipe et chaque statistique, la somme est faite sur tous les matchs de l'automne
for equipe in equipes:
    for stat in off_home_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.HomeTeam==equipe,stat[2:]])
    for stat in off_away_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.AwayTeam==equipe,stat[2:]])

stats_automne
```

```python
#c = conceded
def_home_stat_names = ['c_FTHG', 'c_HTHG'] + ['c_H' + stat_name for stat_name in additional_stat_names]
def_away_stat_names = ['c_FTAG', 'c_HTAG'] + ['c_A' + stat_name for stat_name in additional_stat_names]
def_stat_names = def_home_stat_names +  def_away_stat_names
def_stat_names
```

```python
#Ajouts des statistiques defensives de chaque équipe
for equipe in equipes:
    for stat in def_home_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.AwayTeam==equipe,stat[2:]])
    for stat in def_away_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.HomeTeam==equipe,stat[2:]])

stats_automne
```

```python
home_stat_names = off_home_stat_names + def_home_stat_names
away_stat_names = off_away_stat_names + def_away_stat_names
stat_names = home_stat_names + away_stat_names
```

```python
data2 = data.loc[:,['HomeTeam','AwayTeam','Test group','FTR']]

for stat in home_stat_names:
    data2[stat] = data2.HomeTeam.apply(lambda x: stats_automne.loc[x,stat])
for stat in away_stat_names:
    data2[stat] = data2.AwayTeam.apply(lambda x: stats_automne.loc[x,stat])

data2
```
