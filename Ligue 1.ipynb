# Preprocessing : Homogénéisation du train set et test set (2)

```python
equipes = data.HomeTeam.unique()
```

```python
equipes
```

```python
n_matchs_MS = {}
for i in equipes:
    n_matchs_MS[i] = len(data[(data['HomeTeam']==i)&(data['Test group'] == 0)] + data[(data['AwayTeam']==i)&(data['Test group'] == 0)])
```

```python
n_matchs_MS
```

```python
#Création dictionnaire référençant ensemble des matchs du groupe train n'ayant pas été joué en premiere moitié de saison
matchs_manquants = {}
for i in equipes:
    if n_matchs_MS[i]<19:
        l = list(equipes)
        l.remove(i)
        for j in data[(data['HomeTeam']==i)&(data['Test group']==0)]['AwayTeam']:
            l.remove(j)
        for j in data[(data['AwayTeam']==i)&(data['Test group']==0)]['HomeTeam']:
            l.remove(j)
        matchs_manquants[i] = l
```

```python
matchs_manquants
```

```python
#Assignation des matchs du groupe train au groupe train
for i in matchs_manquants.keys():
    for j in matchs_manquants[i]:
        ref = data[(data['HomeTeam']==i) & (data['AwayTeam']==j)].index[0]
        data.iloc[ref,1] = 0
        matchs_manquants[i].remove(j)
        matchs_manquants[j].remove(i)
```

# Preprocessing : Extraction des variables explicatives (3)

```python
data.columns
```

```python
# S = shots, ST = shots on target, C = corners
additional_stat_names = ['S', 'ST', 'F','C','Y','R']

# décline en Home et Away
# t = taken
off_home_stat_names = ['t_FTHG', 't_HTHG'] + ['t_H' + stat_name for stat_name in additional_stat_names]
off_away_stat_names = ['t_FTAG', 't_HTAG'] + ['t_A' + stat_name for stat_name in additional_stat_names]
off_stat_names = off_home_stat_names +  off_away_stat_names
off_stat_names
```

```python
stats_automne = pd.DataFrame(index = equipes, columns = off_stat_names)

# pour chaque équipe et chaque statistique, la somme est faite sur tous les matchs de l'automne
for equipe in equipes:
    for stat in off_home_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.HomeTeam==equipe,stat[2:]])
    for stat in off_away_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.AwayTeam==equipe,stat[2:]])

stats_automne
```

```python
#c = conceded
def_home_stat_names = ['c_FTHG', 'c_HTHG'] + ['c_H' + stat_name for stat_name in additional_stat_names]
def_away_stat_names = ['c_FTAG', 'c_HTAG'] + ['c_A' + stat_name for stat_name in additional_stat_names]
def_stat_names = def_home_stat_names +  def_away_stat_names
def_stat_names
```

```python
#Ajouts des statistiques defensives de chaque équipe
for equipe in equipes:
    for stat in def_home_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.AwayTeam==equipe,stat[2:]])
    for stat in def_away_stat_names:
        stats_automne.loc[equipe,stat] = np.sum(data_automne.loc[data_automne.HomeTeam==equipe,stat[2:]])

stats_automne
```

```python
#ajout de statistiques rationnelles supplémentaires
# shots_ratio = shots taken/(shots taken + shots conceded)
#shots_accuracy = shots on target/(shots taken + shots on target)
stats_automne['home_shots_ratio'] = stats_automne.loc[:,'t_HS']/(stats_automne.loc[:,'t_HS'] + stats_automne.loc[:,'c_HS'])
stats_automne['away_shots_ratio'] = stats_automne.loc[:,'t_AS']/(stats_automne.loc[:,'t_AS'] + stats_automne.loc[:,'c_AS'])
stats_automne['shots_ratio'] = (stats_automne.loc[:,'t_HS'] + stats_automne.loc[:,'t_AS'])/(stats_automne.loc[:,'t_AS'] + stats_automne.loc[:,'c_AS'] + stats_automne.loc[:,'t_HS'] + stats_automne.loc[:,'c_HS'])
stats_automne['home_shots_accuracy'] = stats_automne.loc[:,'t_HST']/(stats_automne.loc[:,'t_HST'] + stats_automne.loc[:,'t_HS'])
stats_automne['away_shots_accuracy'] = stats_automne.loc[:,'t_AST']/(stats_automne.loc[:,'t_AST'] + stats_automne.loc[:,'t_AS'])
stats_automne['shots_accuracy'] = (stats_automne.loc[:,'t_HST'] + stats_automne.loc[:,'t_AST'])/(stats_automne.loc[:,'t_HST'] + stats_automne.loc[:,'t_HS'] + stats_automne.loc[:,'t_AST'] + stats_automne.loc[:,'t_AS'])
```

```python
home_stat_names = off_home_stat_names + def_home_stat_names + ['home_shots_ratio','home_shots_accuracy']
away_stat_names = off_away_stat_names + def_away_stat_names + ['away_shots_ratio','away_shots_accuracy']
stat_names = home_stat_names + away_stat_names
```

```python
data2 = data.loc[:,['HomeTeam','AwayTeam','Test group','FTR']]

for stat in home_stat_names:
    data2[stat] = data2.HomeTeam.apply(lambda x: stats_automne.loc[x,stat])
for stat in away_stat_names:
    data2[stat] = data2.AwayTeam.apply(lambda x: stats_automne.loc[x,stat])

data2
```

```python
var_expl = stat_names
```

# Preprocessing : création train set et test set (4)

```python
data2_train = data2.loc[data2['Test group']==0,:]
data2_test = data2.loc[data2['Test group']==1,:]
```

```python
#On prend en données d'entrainement les matchs de la première moitié de la saison (plus les 'matchs manquants') et en données d'évaluation les matchs de la 2ème moitié de saison
X_train = data2_train[var_expl].values
y_train = data2_train.FTR.values
X_test = data2_test[var_expl].values
y_test = data2_test.FTR.values
```

# Processing : essaie de quelques modèles (1)

```python
#Pour chaque modèle, on prendra les meilleurs hyperparametres
from sklearn.model_selection import GridSearchCV
```

```python
#Essaie d'un modèle de regression logisitque
from sklearn.linear_model import LogisticRegressionCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
logreg = make_pipeline(StandardScaler(),LogisticRegressionCV(cv=5,verbose=True,n_jobs=-1,Cs=[10**i for i in range(-10,10)],max_iter=20000))
logreg.fit(X_train,y_train)
print('Score données entrainement: ',logreg.score(X_train,y_train), 
      'Score données test : ',logreg.score(X_test,y_test))
```

```python
#Essai d'un modèle de k-plus proches voisins
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
print('Score données entrainement: ',knn.score(X_train,y_train), 
      'Score données test : ',knn.score(X_test,y_test))
```

```python
#essai d'un modèle d'arbre de décisions
from sklearn import tree
arbre = tree.DecisionTreeClassifier()
arbre.fit(X_train,y_train)
print('Score données entrainement: ',arbre.score(X_train,y_train), 
      'Score données test : ',arbre.score(X_test,y_test))
```

```python
#essai d'un modèle de Random Forest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100,bootstrap=True,max_depth=5)
RF.fit(X_train,y_train)
print('Score données entrainement : ',RF.score(X_train,y_train))
print('Score données test : ',RF.score(X_test,y_test))
```

# Processing : voting (2)

```python
#Combinaison des différents modèles par méthode de voting
from sklearn.ensemble import VotingClassifier
voix = VotingClassifier([('k plus proches voisins',knn),
                           ('regressions logisitique', logreg),
                           ('arbre de décision',arbre),
                        ('Random Forest',RF)],
                        voting='hard')
voix.fit(X_train,y_train)
print('Score données entrainement : ',voix.score(X_train,y_train))
print('Score données test : ',voix.score(X_test,y_test))
```
