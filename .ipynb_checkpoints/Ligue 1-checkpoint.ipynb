```python
import os
import pandas as pd
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
import csv
import seaborn as sns
```

```python
data = pd.read_csv('ligue1_2019.csv',index_col='Date',parse_dates=True)
```

```python
data.head()
```

```python
data = data.drop(data.columns[21:],axis = 1)
```

```python
data = data.drop('Div',axis = 1)
```

```python
data.columns
```

```python
data
```

```python
#encodage du resultat Ã  la mi temps
data.loc[:,'HTR'] = data.loc[:,'HTR'].map({'H':1,'D':0,'A':-1})
```

```python
#Encodage du resultat de fin de match
data.loc[:,'FTR'] = data.loc[:,'FTR'].map({'H':1,'D':0,'A':-1})
```

```python
equipes = data.HomeTeam.unique()
```

```python
equipes
```

```python
n_matchs_MS = {}
for i in equipes:
    n_matchs_MS[i] = len(data[data['HomeTeam']==i].loc['2018'] + data[data['AwayTeam']==i].loc['2018'])
    
matchs_manquants = {}
for i in equipes:
    if n_matchs_MS[i]<19:
        l = list(equipes)
        l.remove(i)
        for j in data[data['HomeTeam']==i].loc['2018']['AwayTeam']:
            l.remove(j)
        for j in data[data['AwayTeam']==i].loc['2018']['HomeTeam']:
            l.remove(j)
        matchs_manquants[i] = l
```

```python
X_train = data.loc['2018']
X_test = data.loc['2019']
```

```python
for i in matchs_manquants.keys():
    for j in matchs_manquants[i]:
        match_manquant = data[(data['HomeTeam']==i) & (data['AwayTeam']==j)]
        X_train = X_train.append(match_manquant)
        matchs_manquants[i].remove(j)
        matchs_manquants[j].remove(i)
```

```python
domicile = X_train.groupby('HomeTeam')[['FTHG','FTAG','FTR']].mean()
exterieur = X_train.groupby('AwayTeam')[['FTAG','FTHG','FTR']].mean()
print(domicile,exterieur)
```

```python
def categorie_resultat(FTR):
    if FTR <= -0.5:
        return(0)
    elif (FTR>-0.5)&(FTR<=0):
        return(1)
    elif (FTR<=0.5)&(FTR>0):
        return(2)
    else:
        return(3)

for i in equipes:
    domicile.loc[i,'FTR'] = categorie_resultat(domicile.loc[i,'FTR'])
    exterieur.loc[i,'FTR'] = categorie_resultat(exterieur.loc[i,'FTR'])
```

```python
sns.catplot(x = 'FTHG', y = 'FTAG', data = domicile,hue = 'FTR')
sns.catplot(x = 'FTHG', y = 'FTAG', data = exterieur,hue = 'FTR')
```
